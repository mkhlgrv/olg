# Population

```python
import matplotlib.pyplot as plt
import os
import json
import os
import pandas as pd
from dotenv import load_dotenv
load_dotenv()
import numpy as np
project_dir = os.getenv('project_dir')
import pickle
```

```python
demo_scenarios = ('High variant', 'Medium variant', 'Low variant')
```

```python
demography = {}
for scenario in demo_scenarios:
    with open(os.path.join(project_dir,'assets', f'demography_{scenario.lower().split(" ")[0]}.pickle'), 'rb') as f:
        demography[scenario] = pickle.load(f)
```

```python
for scenario in demo_scenarios:
    plt.plot(range(2018, 2118),demography[scenario]['population'].sum(axis=0).sum(axis=0)[:100], label = scenario)
plt.title('Population')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize = (10,5))
for scenario, ax in zip(demo_scenarios, axes):
    ax.plot(range(2018, 2118),demography[scenario]['deaths'].sum(axis=0).sum(axis=0)[:100], label = f"deaths")
    ax.plot(range(2018, 2118),demography[scenario]['births'].sum(axis=0).sum(axis=0)[:100], label = f"births")
    ax.plot(range(2018, 2118),demography[scenario]['migration'].sum(axis=0).sum(axis=0)[:100], label = f"migration")
    ax.set_title(scenario)
ax.legend()
plt.show()
```

# Oil

```python
oil_scenarios = ('high', 'mid', 'low', 'price_cap')
```

```python
oil = {}
for scenario in oil_scenarios:
    with open(os.path.join(project_dir,'assets', f'oil_{scenario}.pickle'), 'rb') as f:
        oil[scenario] = pickle.load(f)
```

```python
oil[scenario].keys()
```

```python
fig, axes = plt.subplots(2, 2, figsize = (8,10))
axes = axes.flatten()
for scenario, ax in zip(oil_scenarios, axes):
    ax.plot(range(2018, 2048),oil[scenario]['price_O'][:30], label = f"price_O")
    ax.plot(range(2018, 2048),oil[scenario]['psi_O'][:30], label = f"psi_O")
    ax2 = ax.twinx()
    ax2.plot(range(2018, 2048),oil[scenario]['Y_O'][:30], label = f"Y_O", color= "black")
    ax.set_title(scenario)
ax.legend(loc='right')
plt.show()
```
